**Pregunta:** Explique cómo funciona SVD (como modelo, no piense en la implementación), y cómo funciona el método `train()` en  `pyreclab`





**Respuesta:**

SVD permite predecir ratings para cada par usuario-elemento. Para ello, teniendo una matriz de usuarios por elementos con los ratings respectivos, SVD es aplicado sobre la matriz para encontrar factores latentes relacionados a los elementos y usuarios por medio de factorización de matrices. Estos factores latentes son inferidos en base a los ratings provistos [1].

En el caso específico de Funk SVD, se utilizan dos matrices U (usuarios por factores latentes) y V^T (factores latentes por elementos) para ir ajustando los valores de estos factores latentes [1].

Respecto al método train() de PyRecLab, dado un número arbitrario de factores, el método busca ajustar los valores de las matrices U y V^T tomando los ratings que han sido provistos por los usuarios y calculando los valores que deberían adoptar los factores latentes para cada usuario y elemento. Esto es realizado por una cantidad dada de iteraciones y limitando el ajuste realizado en base a la tasa de aprendizaje, a modo de evitar over fitting.

Vinceslas, Medhy. (2019). How Does the Funk Singular Value Decomposition Algorithm work in Recommendation Engines? Data Driven Investor, Medium.

## SVD

**Paso 5**: Tome un usuario, cualquiera, del dataset y reporte parte de sus gustos explícitamente (es decir, nombre de las películas, género, etc., no id). Para este mismo usuario, revise explícitamente las recomendaciones al llegar a la parametrización que reporta las mejores métricas.

```python
user_id = 2
user_df = df_train[df_train['userid'] == user_id]

df_items.loc[user_df['itemid'].values]
```

```python
# Definicion de objeto svd
svd = pyreclab.SVD(dataset='u2.base',
                   dlmchar=b'\t',
                   header=False,
                   usercol=0,
                   itemcol=1,
                   ratingcol=2)

# Entrenamiento del modelo
svd.train(factors=100, maxiter=100, lr=0.01, lamb=0.1)
```

```python
# Testing de predicciones
predlist, mae, rmse = svd.test(input_file='u2.test',
                               dlmchar=b'\t',
                               header=False,
                               usercol=0,
                               itemcol=1,
                               ratingcol=2)

print('MAE: {}\nRMSE: {}'.format(mae, rmse))
```

```python
# Testing de recomendaciones
top_n = 20

recommendList, maprec, ndcg = svd.testrec(input_file='u2.test',
                                          dlmchar=b'\t',
                                          header=False,
                                          usercol=0,
                                          itemcol=1,
                                          ratingcol=2,
                                          topn=top_n,
                                          relevance_threshold=2,
                                          includeRated=False)

print('MAP: {}\nNDCG@{}: {}'.format(maprec, top_n, ndcg))
```

```python
# Calcular las recomendaciones para el usuario escogido
ranking = [int(r) for r in svd.recommend(str(user_id), top_n, includeRated=False)]
print('Recommendation for user {}: {}'.format(user_id, ranking))
```

```python
# Ver explicitamente las recomendaciones para un usuario
df_items.loc[ranking]
```

**Respuesta Pasos 6 a 8:**

Primero se evaluarán 5 valores distintos para número de factores, número de iteraciones máxima, tasa de aprendizaje y lambda, respectivamente, manteniendo los valores dados por defecto para los otros 3 valores cada vez. Luego, se generarán recomendaciones para el usuario utilizando los mejores valores observados para cada parámetro.

```python
#Factores

factor_values = [50, 100, 150, 200, 250]
mae_values = []
rmse_values = []

for f in factor_values:
  svd.train(factors=f, maxiter=100, lr=0.01, lamb=0.1)
  
  predlist, mae, rmse = svd.test(input_file='u2.test',
                                 dlmchar=b'\t',
                                 header=False,
                                 usercol=0,
                                 itemcol=1,
                                 ratingcol=2)
  
  mae_values.append(mae)
  rmse_values.append(rmse)

print(mae_values)
print(rmse_values)

plt.plot(factor_values, mae_values, 'r-')
plt.show()

plt.plot(factor_values, rmse_values, 'bo-')
plt.show()
```

```python
#Iteraciones

maxiter_values = [50, 100, 150, 200, 250]
mae_values = []
rmse_values = []

for i in maxiter_values:
  svd.train(factors=100, maxiter=i, lr=0.01, lamb=0.1)
  
  predlist, mae, rmse = svd.test(input_file='u2.test',
                                 dlmchar=b'\t',
                                 header=False,
                                 usercol=0,
                                 itemcol=1,
                                 ratingcol=2)
  
  mae_values.append(mae)
  rmse_values.append(rmse)

print(mae_values)
print(rmse_values)

plt.plot(maxiter_values, mae_values, 'r-')
plt.show()

plt.plot(maxiter_values, rmse_values, 'bo-')
plt.show()
```

```python
#Tasa de Aprendizaje

lr_values = [0.001, 0.005, 0.01, 0.05, 0.1]
mae_values = []
rmse_values = []

for lr in lr_values:
  svd.train(factors=100, maxiter=100, lr=lr, lamb=0.1)
  
  predlist, mae, rmse = svd.test(input_file='u2.test',
                                 dlmchar=b'\t',
                                 header=False,
                                 usercol=0,
                                 itemcol=1,
                                 ratingcol=2)
  
  mae_values.append(mae)
  rmse_values.append(rmse)

print(mae_values)
print(rmse_values)

plt.plot(lr_values, mae_values, 'r-')
plt.show()

plt.plot(lr_values, rmse_values, 'bo-')
plt.show()
```

```python
#Lambda

lambda_values = [0.05, 0.075, 0.1, 0.5, 0.75]
mae_values = []
rmse_values = []

for l in lambda_values:
  svd.train(factors=100, maxiter=100, lr=0.01, lamb=l)
  
  predlist, mae, rmse = svd.test(input_file='u2.test',
                                 dlmchar=b'\t',
                                 header=False,
                                 usercol=0,
                                 itemcol=1,
                                 ratingcol=2)
  
  mae_values.append(mae)
  rmse_values.append(rmse)

print(mae_values)
print(rmse_values)

plt.plot(lambda_values, mae_values, 'r-')
plt.show()

plt.plot(lambda_values, rmse_values, 'bo-')
plt.show()
```

En base a los resultados, se puede ver que:
* Una mayor cantidad de factores permite reducir el error. Sin embargo, luego de 200 factores, pareciera ser que se produce over fitting pues aumenta el error.
* Con 50 iteraciones se obtiene un error menor que haciendo más iteraciones. Esto daría a entender que realizar demasiadas iteraciones podría provocar over fitting.
* Con una tasa de aprendizaje de 0,005 se consigue el menor error, lo que implicaría el mejor descenso de gradiente entre las tasas de aprendizaje evaluadas.
* Un lambda de 0,1 provee el error más bajo, siendo así el más recomendable entre los evaluados.
