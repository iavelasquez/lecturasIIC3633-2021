```python
import pandas as pd
import numpy as np
import implicit
import scipy.sparse as sparse
```

```python
columns = ['movieid', 'title', 'release_date', 'video_release_date', \
           'IMDb_URL', 'unknown', 'Action', 'Adventure', 'Animation', \
           'Children', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', \
           'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', \
           'Thriller', 'War', 'Western']
```

```python
# Primero creamos el dataframe con los datos
df_train = pd.read_csv('u2.base',
                         sep='\t',
                         names=['userid', 'itemid', 'rating', 'timestamp'],
                         header=None)

# rating >= 3 , relevante (1) y rating menor a 3 es no relevante (0)
df_train.rating = [1 if x >=3 else 0 for x in df_train.rating ]
```

```python
# Cargamos el dataset con los items
df_items = pd.read_csv('u.item',
                        sep='|',
                        index_col=0,
                        names = columns,
                        header=None, 
                        encoding='latin-1')
```

```python
# Cargamos el dataset de testing
df_test = pd.read_csv('u2.test',
                      sep='\t',
                      names=['userid', 'itemid', 'rating', 'timestamp'],
                      header=None)


# rating >= 3 es relevante (1) y rating menor a 3 es no relevante (0) 
df_test.rating = [1 if x >=3 else 0 for x in df_test.rating ]


user_items_test = {}

for row in df_test.itertuples():
    if row[1] not in user_items_test:
        user_items_test[row[1]] = []
        
    user_items_test[row[1]].append(row[2])
```

### Métricas

```python
# Definicion de métricas (No editar)
# Obtenido de https://gist.github.com/bwhite/3726239

def precision_at_k(r, k):
    assert k >= 1
    r = np.asarray(r)[:k] != 0
    if r.size != k:
        raise ValueError('Relevance score length < k')
    return np.mean(r)

def average_precision(r):
    r = np.asarray(r) != 0
    out = [precision_at_k(r, k + 1) for k in range(r.size) if r[k]]
    if not out:
        return 0.
    return np.mean(out)

def mean_average_precision(rs):
    return np.mean([average_precision(r) for r in rs])
  
def dcg_at_k(r, k):
    r = np.asfarray(r)[:k]
    if r.size:
        return np.sum(np.subtract(np.power(2, r), 1) / np.log2(np.arange(2, r.size + 2)))
    return 0.


def ndcg_at_k(r, k):
    idcg = dcg_at_k(sorted(r, reverse=True), k)

    if not idcg:
        return 0.
    return dcg_at_k(r, k) / idcg
```

### Preprocesamiento de los datos a formato sparse

```python
user_items = {}
itemset = set()

for row in df_train.itertuples():
    if row[1] not in user_items:
        user_items[row[1]] = []
        
    user_items[row[1]].append(row[2])
    itemset.add(row[2])

itemset = np.sort(list(itemset))

sparse_matrix = np.zeros((len(user_items), len(itemset)))

for i, items in enumerate(user_items.values()):
    sparse_matrix[i] = np.isin(itemset, items, assume_unique=True).astype(int)
    
matrix = sparse.csr_matrix(sparse_matrix.T)

user_ids = {key: i for i, key in enumerate(user_items.keys())}
user_item_matrix = matrix.T.tocsr()
```

```python
def evaluate_model(model, n):
  mean_map = 0.
  mean_ndcg = 0.
  for u in user_items_test.keys():
    rec = [t[0] for t in model.recommend(u, user_item_matrix, n)]
    rel_vector = [np.isin(user_items_test[u], rec, assume_unique=True).astype(int)]
    mean_map += mean_average_precision(rel_vector)
    mean_ndcg += ndcg_at_k(rel_vector, n)

  mean_map /= len(user_items_test)
  mean_ndcg /= len(user_items_test)
  
  return mean_map, mean_ndcg
```

```python
def show_recommendations(model, user, n):
  recommendations = [t[0] for t in model.recommend(user, user_item_matrix, n)]
  return df_items.loc[recommendations]['title']
```

```python
def show_similar_movies(model, item, n=10):
  sim_items = [t[0] for t in model.similar_items(item, n)]
  return df_items.loc[sim_items]['title']
```

## ALS (Implicit Feedback)

**Pregunta 1:** Explique brevemente cómo funciona el algoritmo ALS.

**Respuesta:**

ALS factoriza una matriz de ratings de usuarios por elementos en dos matrices incluyendo los factores latentes de los usuarios y los elementos, respectivamente. Utilizando una función de costo, ALS completa los valores en ambas matrices por medio de factorización. Esta función de costo tiene la gracia de que su valor se ve reducido en cada paso realizado por el algoritmo, actuando esto como el método de optimización del algoritmo.

```python
# Definimos y entrenamos el modelo con optimización ALS
model_als = implicit.als.AlternatingLeastSquares(factors=100, iterations=10, use_gpu=False)
model_als.fit(matrix)
```

Ejemplo de recomendación y búsqueda de items similares con los factores latentes ya entrenados:

```python
show_recommendations(model_als, user=77, n=10)
```

```python
maprec, ndcg = evaluate_model(model_als, n=10)
print('map: {}\nndcg: {}'.format(maprec, ndcg))
```

**Pregunta 2:** Pruebe distintos valores para los parámetros de ALS y muestre gráficos de cómo se ven afectadas las métricas recién mostradas.

**Respuesta:**

Se han analizado todas las combinaciones para factores = [50, 100, 150, 200, 250], e iteraciones = [5, 10, 25, 50, 100], como se presenta a continuación.

Tanto para MAP como para nDCG@10, el mejor resultado obtenido fue para factores = 100, e iteraciones = 5.

```python
import matplotlib.pyplot as plt
```

```python
factor_values = [50, 100, 150, 200, 250]
iterations_values = [5, 10, 25, 50, 100]

map_values = []
ndcg_values = []

for iteration in iterations_values:
  iteration_map_values = []
  iteration_ndcg_values = []
  
  for value in factor_values:
    model = implicit.als.AlternatingLeastSquares(factors=value, iterations=iteration, use_gpu=False)
    model.fit(matrix)

    maprec, ndcg = evaluate_model(model, n=10)

    iteration_map_values.append(maprec)
    iteration_ndcg_values.append(ndcg)

  map_values.append(iteration_map_values)
  ndcg_values.append(iteration_ndcg_values)

for iteration, values in zip(iterations_values, map_values):
  plt.plot(factor_values, values, label = str(iteration) + " iterations")
plt.xlabel("Factors")
plt.ylabel("MAP")
plt.grid()
plt.legend()
plt.show()

for iteration, values in zip(iterations_values, ndcg_values):
  plt.plot(factor_values, values, label = str(iteration) + " iterations")
plt.xlabel("Factors")
plt.ylabel("nDCG@10")
plt.grid()
plt.legend()
plt.show()
```

## BPR

**Pregunta 3:** Explique con sus palabras la intuición del framework BPR.

**Respuesta:**

BPR es una función de pérdida basada en la suposición de que un usuario prefiere los elementos que ya ha consumido por sobre aquellos que aun no ha consumido. Así, el algoritmo busca maximizar la probabilidad de que, dados dos elementos *i* y *j*, el usuario prefiera el elemento *i* sobre el elemento *j*.

```python
# Definimos y entrenamos el modelo de implicit feedback utilizando optimizacion BPR
model_bpr = implicit.bpr.BayesianPersonalizedRanking(factors=400, iterations=40, use_gpu=False)
model_bpr.fit(matrix)
```

Ejemplo de recomendación y búsqueda de items similares con los factores latentes ya entrenados:

```python
show_recommendations(model_bpr, user=77, n=10)
```

```python
show_similar_movies(model_bpr, item=171, n=10)
```

```python
maprec, ndcg = evaluate_model(model_bpr, n=10)
print('map: {}\nndcg: {}'.format(maprec, ndcg))
```

**Pregunta 4:** Pruebe distintos valores para los parámetros de BPR y muestre gráficos de cómo se ven afectadas las métricas de ranking (nDCG@10 y MAP) recién mostradas.

**Respuesta:**

Se han analizado todas las combinaciones para factores = [50, 100, 150, 200, 250], e iteraciones = [5, 10, 25, 50, 100], como se presenta a continuación.

Tanto para MAP como para nDCG@10, el mejor resultado se obtuvo con iteraciones = 5. Sin embargo, el mayor MAP se obtuvo para factores = 150, y el mayor nDCG@10 se obtuvo para factores = 250. Se resalta, además, que con iteraciones = 25, y factores = 150, se obtiene un nDCG@10 bastante cercano al anterior.

También es posible observar que, tanto con MAP como nDCG@10, los valores obtenidos para iteraciones = 100 son los más bajos, independiente de la cantidad de factores. Se presume que esto se debe a que ocurre over fitting al realizar demasiadas iteraciones.

```python
factor_values = [50, 100, 150, 200, 250]
iterations_values = [5, 10, 25, 50, 100]

map_values = []
ndcg_values = []

for iteration in iterations_values:
  iteration_map_values = []
  iteration_ndcg_values = []
  
  for value in factor_values:
    model = implicit.bpr.BayesianPersonalizedRanking(factors=value, iterations=iteration, use_gpu=False)
    model.fit(matrix)

    maprec, ndcg = evaluate_model(model, n=10)

    iteration_map_values.append(maprec)
    iteration_ndcg_values.append(ndcg)

  map_values.append(iteration_map_values)
  ndcg_values.append(iteration_ndcg_values)

for iteration, values in zip(iterations_values, map_values):
  plt.plot(factor_values, values, label = str(iteration) + " iterations")
plt.xlabel("Factors")
plt.ylabel("MAP")
plt.grid()
plt.legend()
plt.show()

for iteration, values in zip(iterations_values, ndcg_values):
  plt.plot(factor_values, values, label = str(iteration) + " iterations")
plt.xlabel("Factors")
plt.ylabel("nDCG@10")
plt.grid()
plt.legend()
plt.show()
```
